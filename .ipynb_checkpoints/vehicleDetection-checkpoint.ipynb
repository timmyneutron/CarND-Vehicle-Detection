## Histograms of Oriented Gradients (HOG) Features

I calculated HOG features for images using the `sklearn.feature.hog` method, using the following parameters:

  - 9 bins for gradient orientations, which is shown [here](http://lear.inrialpes.fr/people/triggs/pubs/Dalal-cvpr05.pdf) to be the optimal number for a similar classification problem.
  - 8x8 pixel per cell, which is an appropriate number given the size of the features we're looking for in the image (~8x8 pixels)
  - 2x2 cells per block, which gives a good local normalization for gradients in each block

Also, HOG features capture gradients, so I converted the original RGB images to YCrCb - YCrCb puts most of the visibility information (luminance) in the Y channel, and the color information in the Cr (chroma red) and Cb (chroma blue) channels, while maintaining the original shape of the image in all three channels. This helps to emphasize the gradients in the image.

Examples of car and not-car images show below with their respective HOG images, in which gradient features are clearly visible.

```python
def get_hog_features(img, orient, pix_per_cell, cell_per_block, 
                        vis=False, feature_vec=True):
    '''Function for getting the HOG features for an image'''
    
    # if vis is True, return features and image
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block), 
                                  transform_sqrt=True, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image

    # if vis is False, return just the features
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), 
                       transform_sqrt=True, 
                       visualise=vis, feature_vector=feature_vec)
        
        return features

# define HOG parameters
orient = 9
pix_per_cell = 8
cell_per_block = 2

# plot original and HOG images
fig = plt.figure(figsize=(7, 7))

car_img = mpimg.imread(np.random.choice(cars))
car_lum = cv2.cvtColor(car_img, cv2.COLOR_RGB2YCrCb)[:,:,0]
car_hog_features, car_hog_img = get_hog_features(car_lum, orient, pix_per_cell, cell_per_block, 
                    vis=True, feature_vec=True)

fig.add_subplot(2, 2, 1)
plt.imshow(car_img)
plt.title("Car Image")

fig.add_subplot(2, 2, 2)
plt.imshow(car_hog_img, cmap="gray")
plt.title("Car Hog Image")

notcar_img = mpimg.imread(np.random.choice(notcars))
notcar_lum = cv2.cvtColor(notcar_img, cv2.COLOR_RGB2YCrCb)[:,:,0]
notcar_hog_features, notcar_hog_img = get_hog_features(notcar_lum, orient, pix_per_cell, cell_per_block, 
                    vis=True, feature_vec=True)

fig.add_subplot(2, 2, 3)
plt.imshow(notcar_img)
plt.title("Not Car Image")

fig.add_subplot(2, 2, 4)
plt.imshow(notcar_hog_img, cmap="gray")
plt.title("Not Car Hog Image")
```

## Spatial Binning

I converted the 64x64 training images to 16x16 images, and unraveled those to extract spatial binning features. Converting to 16x16 reduces the feature vector size, while still maintaining the relevant features of the images (show below).

```python
def bin_spatial(img, size=(16, 16)):
    '''Function for getting the spatial binning features of an image'''
    features = cv2.resize(img, size).ravel() 
    
    return features

# plot original and reduced sized images
car_img = mpimg.imread(np.random.choice(cars))
notcar_img = mpimg.imread(np.random.choice(notcars))

car_img_res = cv2.resize(car_img, (16, 16))
notcar_img_res = cv2.resize(notcar_img, (16, 16))

fig = plt.figure(figsize=(7, 7))

fig.add_subplot(2, 2, 1)
plt.imshow(car_img)
plt.title("Car Image")

fig.add_subplot(2, 2, 2)
plt.imshow(car_img_res, cmap="gray")
plt.title("Reduced Size Car Image")

fig.add_subplot(2, 2, 3)
plt.imshow(notcar_img)
plt.title("Not Car Image")

fig.add_subplot(2, 2, 4)
plt.imshow(notcar_img_res, cmap="gray")
plt.title("Reduced Size Not Car Image")
```

## Color Histograms

I took the color histograms for each channel in each image, using 32 bins for the histograms. This captures information related to the prevalence of certain colors in an image, without regard to position within the image.

Color histograms (in YCrCb) for car and not-car images show below.

```python
def color_hist(img, nbins=32):
    '''Function for getting the color histogram features of an image'''
    
    channel1_hist = np.histogram(img[:,:,0], bins=nbins)
    channel2_hist = np.histogram(img[:,:,1], bins=nbins)
    channel3_hist = np.histogram(img[:,:,2], bins=nbins)

    hist_features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))
    
    return hist_features

fig = plt.figure(figsize=(7, 7))

car_img = mpimg.imread(np.random.choice(cars))
notcar_img = mpimg.imread(np.random.choice(notcars))

car_hist = color_hist(car_img, nbins=32)
notcar_hist = color_hist(notcar_img, nbins=32)

fig.add_subplot(2, 1, 1)
plt.plot(car_hist)
plt.title("Car Histogram")

fig.add_subplot(2, 1, 2)
plt.plot(notcar_hist)
plt.title("Not Car Histogram")
```

```python
 def extract_features(imgs, color_space="RGB",
                        spatial_size=(32, 32),
                        hist_bins=32, orient=9, 
                        pix_per_cell=8, cell_per_block=2, hog_channel="ALL",
                        spatial_feat=True, hist_feat=True, hog_feat=True):
    '''Function for getting all features for a list of images.
       Takes in a list of image paths and returns a list of feature vectors.'''
    
    # initialize list of feature vectors
    features = []
    
    # loop through list of image paths
    for file in imgs:
        # initialize list of features for single image
        file_features = []
        
        # read image
        # note: images are png files, so mpimg.imread returns values as floats from 0 to 1, NOT integers from 255
        image = mpimg.imread(file)

        # convert to specified color space
        if color_space != 'RGB':
            conversion = "cv2.COLOR_RGB2{}".format(color_space)
            feature_image = cv2.cvtColor(image, eval(conversion))
        else:
            feature_image = np.copy(image)      

        # extract spatial features
        if spatial_feat == True:                             
            spatial_features = bin_spatial(feature_image, size=spatial_size)
            file_features.append(spatial_features)
            
        # extract color histogram features
        if hist_feat == True:
            hist_features = color_hist(feature_image, nbins=hist_bins)
            file_features.append(hist_features)
        
        # extract hog features
        if hog_feat == True:
            
            if hog_channel == 'ALL':
                hog_features = []
                for channel in range(feature_image.shape[2]):
                    hog_features.append(get_hog_features(feature_image[:,:,channel], 
                                        orient, pix_per_cell, cell_per_block, 
                                        vis=False, feature_vec=True))
                hog_features = np.ravel(hog_features)        
            else:
                hog_features = get_hog_features(feature_image[:,:,hog_channel], orient, 
                            pix_per_cell, cell_per_block, vis=False, feature_vec=True)
           
            file_features.append(hog_features)
            
        # append image feature vector to list of feature vectors
        features.append(np.concatenate(file_features))
    
    return features
```

## Training the classifier

I extracted spatial, color histogram, and HOG features. This gives a feature vector of length 6156 features, so I chose to train a LinearSVC classifier on the data. This was because a) support vector machines are robust to data with large numbers of features, and b) once trained, prediction is relatively fast, and the classifier needs to be able to work in real time.

I chose to convert images to YCrCb color space, because (as previously mentioned) it helps in capturing gradients while maintaining the shape of the image in all 3 channels. I confirmed this by training the classifier with different color spaces, and the YCrCb gave the highest classification accuracy, which was over 99%.

I also tested different values of the C parameter for the LinearSVC - the C parameter is a penalty for misclassification, and generally corresponds the bias/variance tradeoff for classification, where a lower C value gives a higher bias model, and a higher C value gives a higher variance model. From this, I found a C value of 10 gave the highest classification accuracy.

```python
# specify random sample size
# if sample_size is None, use whole training set
sample_size = None

if sample_size != None:
    cars = shuffle(cars)[:sample_size]
    notcars = shuffle(notcars)[:sample_size]

# specify parameters for feature extraction

## types of features to extract
spatial_feat = True
hist_feat = True
hog_feat = True

## color space
color_space = "YCrCb"

## HOG parameters
orient = 9
pix_per_cell = 8
cell_per_block = 2
hog_channel = "ALL"

## spatial binning size
spatial_size = (16, 16)

## number of color histogram bins
hist_bins = 32

# start timer
t0 = time.time()

# extract features for car examples
car_features = extract_features(cars, color_space=color_space,      
                        spatial_size=spatial_size, hist_bins=hist_bins, 
                        orient=orient, pix_per_cell=pix_per_cell, 
                        cell_per_block=cell_per_block, 
                        hog_channel=hog_channel, spatial_feat=spatial_feat, 
                        hist_feat=hist_feat, hog_feat=hog_feat)

# extract features for notcar examples
notcar_features = extract_features(notcars, color_space=color_space, 
                        spatial_size=spatial_size, hist_bins=hist_bins, 
                        orient=orient, pix_per_cell=pix_per_cell, 
                        cell_per_block=cell_per_block, 
                        hog_channel=hog_channel, spatial_feat=spatial_feat, 
                        hist_feat=hist_feat, hog_feat=hog_feat)

# record time to extract features
t1 = time.time()

# scale features using StandardScaler
X = np.vstack((car_features, notcar_features)).astype(np.float64)                        
X_scaler = StandardScaler().fit(X)
scaled_X = X_scaler.transform(X)

# define the labels vector
y = np.hstack((np.ones(len(car_features)), np.zeros(len(notcar_features))))

# split up data into randomized training and test sets
rand_state = np.random.randint(0, 100)
X_train, X_test, y_train, y_test = train_test_split(
    scaled_X, y, test_size=0.2, random_state=rand_state)

# record time to scale and split data
t2 = time.time()

# initialize a linear SVC classifier
svc = LinearSVC(C=10.0)

# fit the classifier to the training data
svc.fit(X_train, y_train)

# record time to train the classifier
t3 = time.time()

print('Using:',orient,'orientations',pix_per_cell,
    'pixels per cell and', cell_per_block,'cells per block')
print('Feature vector length:', len(car_features[0]))
print('Test Accuracy of SVC = ', round(svc.score(X_test, y_test), 4))
print()
print("Time to extract features: {} seconds".format(round(t1-t0, 2)))
print("Time to transform vectors: {} seconds".format(round(t2-t1, 2)))
print("Time to train SVC: {} seconds".format(round(t3-t2, 2)))
```

## Finding cars in a video frame

Once the classifier was trained, the next step was to find cars in test video frames. The simplest way to do this is to define moving windows of different sizes to scan the region where cars are expected to be found (the bottom half of the image), extract features in the same way that images were extracted, and use the classifier to predict whether or not that window contains a car.

One drawback to this approach is that finding HOG features is computationally expensive, so instead, HOG features are calculated once for the whole image, and then those features are stepped through to find HOG features. Different window sizes are simulated by resizing the whole image before finding HOG features.

Spatial binning and color histogram features are found by correlating the location of the window in the HOG image with the location of the window in the original pixel image.

I used windows of size 64x64 pixels and 96x96 pixels, since these two combined consistently found the cars while generally avoiding false positives.

## Heat Mapping and Labeling

I used heat mapping and thresholding to identify true positives and eliminate false positives. Each positive window adds +1 "heat" to all its pixels in a heat map image, and once all windows are cycled through, any pixels with a heat below a certain threshold are eliminated. This is effective in removing false positives, as well as separating individual cars into contiguous regions of high heat.

Once thresholding was applied, I used `scipy.ndimage.measurements.label` to label all the contiguous regions on the heat map (which correspond to individual cars), and drew bounding boxes around each region. I also added the label number on top of each car to show the continuity of cars between frames, and to show how many regions are identified in the frame.

Box images, heat maps, and labeled bounding box images show below.

```python
def find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell, cell_per_block, cells_per_step, spatial_size, hist_bins,
             color_space="RGB"):
    '''Function for finding cars in a single video frame.
       Extracts HOG features from whole image once for the sake of efficiency and
           moves window across those features to find HOG features.
       Extracts spatial binning and histogram features from each individual window.
       Makes a prediction using the trained classifier, and if the prediction is positive,
           appends the window corners to a list of hot windows.
       Returns list of hot windows.
    '''
    
    # copy of original image to draw on
    draw_img = np.copy(img)
    
    # Classifier was trained on image data ranging from 0 to 1,
    # but pipeline gets image data from 0 to 255, so need to convert
    img = img.astype(np.float32)/255
    
    # define the region of the image to search for cars
    img_tosearch = img[ystart:ystop,:,:]
    
    # convert to specified color space
    if color_space != "RGB":
        conversion = "cv2.COLOR_RGB2{}".format(color_space)
        ctrans_tosearch = cv2.cvtColor(img_tosearch, eval(conversion))
    else:
        ctrans_tosearch = np.copy(img_tosearch)
    
    # resize window to specified size
    # allows function to search for cars with different-sized windows
    if scale != 1:
        imshape = ctrans_tosearch.shape
        ctrans_tosearch = cv2.resize(ctrans_tosearch, (np.int(imshape[1]/scale), np.int(imshape[0]/scale)))
    
    # split image into channels
    ch1 = ctrans_tosearch[:,:,0]
    ch2 = ctrans_tosearch[:,:,1]
    ch3 = ctrans_tosearch[:,:,2]

    # define number of HOG blocks in x and y directions for HOG feature extraction
    nxblocks = (ch1.shape[1] // pix_per_cell)-1
    nyblocks = (ch1.shape[0] // pix_per_cell)-1 
    
    # number of HOG features per window
    nfeat_per_block = orient*cell_per_block**2
    
    # set window size to 64x64 (same size as training images)
    window = 64
    
    # set number of blocks per window
    nblocks_per_window = (window // pix_per_cell)-1
    
    # set number of steps for windows
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step
    
    # get HOG features for each channel for whole image
    hog1 = get_hog_features(ch1, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog2 = get_hog_features(ch2, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog3 = get_hog_features(ch3, orient, pix_per_cell, cell_per_block, feature_vec=False)
    
    # initialize list of hot windows (windows with a positive prediction)
    hot_windows = []
    
    # loop over number of steps in x and y directions
    for xb in range(nxsteps):
        for yb in range(nysteps):
            
            # set x and y positions on HOG feature grid
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            
            # extract HOG features for the specified window
            hog_feat1 = hog1[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat2 = hog2[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat3 = hog3[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_features = np.hstack((hog_feat1, hog_feat2, hog_feat3))

            # define top corner location on original image in pixels
            xleft = xpos*pix_per_cell
            ytop = ypos*pix_per_cell

            # define pixel image from original image
            subimg = cv2.resize(ctrans_tosearch[ytop:ytop+window, xleft:xleft+window], (64,64))
          
            # get spatial binning features
            spatial_features = bin_spatial(subimg, size=spatial_size)
                
            # get color histogram features
            hist_features = color_hist(subimg, nbins=hist_bins)
            
            # scale features with same scaler used to scale training features
            test_features = X_scaler.transform(np.hstack((spatial_features, hist_features, hog_features)).reshape(1, -1))    
            
            # make prediction using classifier
            test_prediction = svc.predict(test_features)
            
            # if classifier returns a positive prediction, append the corners of the window to
            # list of hot windows
            if test_prediction == 1:
                xbox_left = np.int(xleft*scale)
                ytop_draw = np.int(ytop*scale)
                win_draw = np.int(window*scale)
                
                top_left = (xbox_left, ytop_draw+ystart)
                bottom_right = (xbox_left+win_draw,ytop_draw+win_draw+ystart)
                
                hot_windows.append((top_left, bottom_right))
                
    # return hot windows list                           
    return hot_windows
```

```python
def add_heat(heatmap, bbox_list):
    '''Adds heat to a heat map for each hot window on a frame'''
    
    for box in bbox_list:
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1

    return heatmap
```

```python
def apply_threshold(heatmap, threshold):
    '''Sets heatmap pixels less than or equal to threshold to 0'''
    
    heatmap[heatmap <= threshold] = 0

    return heatmap
```

```python
def draw_labeled_bboxes(img, labels):
    '''Draws boxes on frame around labeled regions'''
    
    # loop through number of contiguous regions
    for car_number in range(1, labels[1]+1):
        
        # find region that corresponds to car_number
        nonzero = (labels[0] == car_number).nonzero()
        
        # find x and y coordinates of nonzero pixels
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        
        # define corners of bounding box using min and max x and y coordinates
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))

        # draw the box on the image
        cv2.rectangle(img, bbox[0], bbox[1], (0,0,255), 6)
        
        # draw label number above box
        cv2.putText(img, str(car_number), bbox[0], cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 3)
        
    # Return the image
    return img
```

```python
# show test images

fig = plt.figure(figsize=(12, 15))

for i in range(1, 7):
    
    # read test images
    img_path = "test_images/test{}.jpg".format(i)
    img = mpimg.imread(img_path)
    
    hot_windows = []
    
    # define search region, scale, and step size
    ystart = 400
    ystop = 656
    scale = 1.0
    cells_per_step = 2
    
    # search for cars and append hot windows to list
    hot_windows = hot_windows + find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell,
                                    cell_per_block, cells_per_step, spatial_size, hist_bins,
                                    color_space=color_space)
    
    # change to different scale
    scale = 1.5
    
    # search for cars again and append hot windows to list
    hot_windows = hot_windows + find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell,
                                    cell_per_block, cells_per_step, spatial_size, hist_bins,
                                    color_space=color_space)
    
    # set heat threshold
    heat_threshold = 1

    box_img = img.copy()
    
    # draw boxes for every hot window
    for (top_left, bottom_right) in hot_windows:
        cv2.rectangle(box_img, top_left, bottom_right,(0,0,255),6)

    # plot box image
    fig.add_subplot(6, 3, 3*i-2)
    plt.imshow(box_img)
    
    if i == 1:
        plt.title("All Boxes")
    
    # add heat to heat map
    heat = np.zeros_like(img[:,:,0]).astype(np.float)
    heat = add_heat(heat, hot_windows)
    
    # apply thresholding to heat map
    heat = apply_threshold(heat, heat_threshold)
    heatmap = np.clip(heat, 0, 255)
    
    # plot heat map
    fig.add_subplot(6, 3, 3*i-1)
    plt.imshow(heatmap)
    
    if i == 1:
        plt.title("Heat Map")
    
    # apply labeling to heat map regions
    labels = label(heatmap)
    
    # draw bounding boxes on heat map regions
    combined_image = draw_labeled_bboxes(img, labels)
    
    # plot bounding box images
    fig.add_subplot(6, 3, 3*i)
    plt.imshow(combined_image)
    
    if i == 1:
        plt.title("Bounding Boxes")
```

## Testing on Video

Since it takes a while to process the output video, I found it efficient to break the project video into 10 second clips for testing. This helped when I only wanted to test changes to the pipeline on certain pieces of the video.

Also in the actual video stream, I smoothed out the predictions by using a moving average of 20 frames for heat mapping before thresholding.

Output video shown below

```python
# Define class to store video stats from frame to frame
class VideoStats():
    def __init__(self):
        
        # number of heat map frames to store for moving average
        self.moving_avg_num = 10
        
        # heat threshold
        self.heat_threshold = 1.8
        
        # array to store heat maps
        self.heatmaps = np.zeros((self.moving_avg_num, 720, 1280))
        
        # initialize frame number
        self.frame_num = 0

# Instantiate VideoStats class
video_stats = VideoStats()

def pipeline(img):
    
    hot_windows = []
    
    # define search region, scale, and step size
    ystart = 400
    ystop = 656
    scale = 1.0
    cells_per_step = 2
    
    # search for cars and append hot windows to list
    hot_windows = hot_windows + find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell,
                                    cell_per_block, cells_per_step, spatial_size, hist_bins,
                                    color_space=color_space)
    
    # change to different scale
    scale = 1.5
    
    # search for cars again and append hot windows to list
    hot_windows = hot_windows + find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell,
                                    cell_per_block, cells_per_step, spatial_size, hist_bins,
                                    color_space=color_space)
    
    # add heat to heat map
    heat = np.zeros_like(img[:,:,0]).astype(np.float)
    heat = add_heat(heat, hot_windows)
    
    video_stats.heatmaps[video_stats.frame_num % video_stats.moving_avg_num] = heat
    
    avg_heat = np.mean(video_stats.heatmaps, axis=0)
    
    # apply thresholding to heat map
    thresh_heat = apply_threshold(avg_heat, video_stats.heat_threshold)
    heatmap = np.clip(thresh_heat, 0, 255)
    
    # apply labeling to heat map regions
    labels = label(heatmap)
    
    # draw bounding boxes on heat map regions
    combined_image = draw_labeled_bboxes(img, labels)
    
    video_stats.frame_num += 1
    
    return combined_image
```

```python
output_video = 'output_video.mp4'
clip1 = VideoFileClip("project_video.mp4")
output_clip = clip1.fl_image(pipeline) #NOTE: this function expects color images!!
%time output_clip.write_videofile(output_video, audio=False)
```

## Discussion/Areas for Improvement

The classifier tracks all cars fairly reliably on the project video. The biggest issue is that it doesn't detect cars just as they are entering the field of vision from the side, and only detects them once they moved quite a bit onto the frame. This is most likely because of the training data used for the classifier - almost all the training images for cars were taken from behind the car, rather than to the side. More training data that fits this case would likely solve this problem.

Another less important issue is that the classifier loses track of individual cars when they overlap, since overlapping heat map regions don't differentiate between cars. This could be solved perhaps with training classifiers for different color cars (perhaps one for black, and one for white), and using more robust logic to keep track of cars when they are occluded.
